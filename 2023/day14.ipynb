```python
# Get data

from aocd import get_data
from collections import Counter
import time
from functools import reduce
from itertools import islice
from functools import cache

# get data
data = get_data(day=14, year=2023)

# data = """O....#....
# O.OO#....#
# .....##...
# OO.#O....O
# .O.....O#.
# O.#..O.#.#
# ..O..#O..O
# .......O..
# #....###..
# #OO..#...."""

#data = [string for string in data.split('\n\n')]

# make North to point left so that we can iterate
# through the list - each element in the list is a 
# vertical column on the provided input
data_transpose = list(map(''.join, zip(*data.split('\n'))))

#print(data_transpose)
```

```python
# Part 1

t0 = time.time()

rock_pos = []
for string in data_transpose:
    row_len = len(string)
    string_bit_start = row_len
    
    string = string.replace('#', '~#~')
    rock_pos_row = []
    #print(string)
    for string_bit in string.split('~'):
        #print(string_bit)
        #print(string_bit_start)
        string_bit_len = len(string_bit)
        rock_pos_row = rock_pos_row + [string_bit_start - x for x in range(Counter(string_bit)['O'])]
        string_bit_start -= string_bit_len
    rock_pos.append(rock_pos_row)
    #print("----")
    
#print(rock_pos)
print(sum([sum(x) for x in rock_pos]))

t1 = time.time()

print(1000*(t1-t0), "msec")
```

```python
# Part 2

def tilt(string_list):
    out_string_list = []
    for string in string_list:
        #print(string)
        new_string_bit = []
        for string_bit in string.split('#'):
            #print(string_bit)
            #print(string_bit_start)
            string_bit_len = len(string_bit)
            if string_bit_len > 0:
                numO = Counter(string_bit)['O']
                new_string_bit = new_string_bit + ['O' * numO + '.' * (string_bit_len - numO)]
            else:
                new_string_bit = new_string_bit + [string_bit]
        out_string_list = out_string_list + ['#'.join(new_string_bit)]
    return out_string_list

# rotates anti-clockwise
def rotate_aw(string_list):
    return list(map(''.join, zip(*reversed(string_list))))

# rotates clockwise
def rotate(string_list):
    return rotate_aw(rotate_aw(rotate_aw(string_list)))

# helps to see steps
def print_string(string_list):
    rotated_list = rotate_aw(string_list)
    print(*[''.join(list(reversed(x))) for x in rotated_list], sep = "\n")
    
def cycle(string_list):
    return rotate(tilt(rotate(tilt(rotate(tilt(rotate(tilt(string_list))))))))
    
def cycle_gen(string_list):
    cycled = rotate(tilt(rotate(tilt(rotate(tilt(rotate(tilt(string_list))))))))
    while True:
        yield cycled
        cycled = rotate(tilt(rotate(tilt(rotate(tilt(rotate(tilt(cycled))))))))
        

def load(string_list):

    load = 0
    for string in string_list:
        row_len = len(string)
        load += sum([row_len-i for i, x in enumerate(string) if x == "O"])
        
    return load

# check part 1
print(load(tilt(data_transpose)))

spinned = cycle_gen(data_transpose)
#cycles = 100000
#spun = next(islice(spinned, cycles-1, None))

for i in range(300):
    spun = next(spinned)
    print(i, load(spun))

# repeating cycle length is 28
# start at 171 when cycle is in full swing
# so we look at the load value at this cycle number:
print(171+(1000000000-171-1)%28)
```
